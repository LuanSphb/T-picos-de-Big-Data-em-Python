```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

# Carregar a tabela do Excel
df = pd.read_excel('Rel_prod.xlsx')

# Eliminar linhas onde a coluna 'Cod_prod' está vazia
df = df.dropna(subset=['Cod_prod'])

# Renomear as colunas
df = df.rename(columns={
    'Cod_prod': 'Codigo do produto',
    'Nome_prod': 'Nome do produto',
    'Qtde_pedida': 'Quantidade',
    'Revendedora': 'Revendedora'
})

# Função para atualizar a tabela com base na revendedora selecionada
def atualizar_tabela(revendedora):
    if revendedora == 'todos':
        df_agrupado = df.groupby(['Codigo do produto', 'Nome do produto']).agg({'Quantidade': 'sum'}).reset_index()
        df_agrupado = df_agrupado.nlargest(30, 'Quantidade')
        return df_agrupado
    else:
        df_filtrado = df[df['Revendedora'] == revendedora]
        df_filtrado = df_filtrado[['Codigo do produto', 'Nome do produto', 'Quantidade']]
        df_filtrado = df_filtrado.nlargest(30, 'Quantidade')
        return df_filtrado

# Função de callback para atualizar a exibição da tabela
def atualizar_tabela_callback(revendedora):
    clear_output(wait=True)
    display(dropdown)
    df_tabela = atualizar_tabela(revendedora)
    html_table = df_tabela.to_html(index=False, classes='table table-striped')  # Converte para HTML sem o índice
    display(HTML(html_table))

# Criar o widget de dropdown
dropdown = widgets.Dropdown(
    options=[('todos', 'todos')] + [(i, i) for i in df['Revendedora'].unique()],
    value='todos',  # Valor inicial do dropdown
    description='Revendedora:',
)

# Conectar o dropdown ao callback
dropdown.observe(lambda change: atualizar_tabela_callback(change['new']), names='value')

# Exibir o dropdown e a tabela inicial
display(dropdown)
atualizar_tabela_callback(dropdown.value)
```
