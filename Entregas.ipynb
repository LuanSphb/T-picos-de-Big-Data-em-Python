```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

# Carregar a tabela do Excel
df = pd.read_excel('Consultores.xlsx')

# Selecionar apenas as colunas desejadas
colunas_desejadas = ['Nome', 'Endereco', 'Bairro', 'Cidade']
df = df[colunas_desejadas]

# Função para atualizar a tabela com base nos filtros aplicados
def atualizar_tabela(bairro_selecionado, nomes_selecionados):
    df_filtrado = df
    if bairro_selecionado != 'todos':
        df_filtrado = df_filtrado[df_filtrado['Bairro'] == bairro_selecionado]
    if nomes_selecionados:
        df_filtrado = df_filtrado[df_filtrado['Nome'].isin(nomes_selecionados)]
    return df_filtrado

# Função de callback para atualizar a exibição da tabela
def atualizar_tabela_callback(_):
    clear_output(wait=True)
    display(dropdown_bairro)
    display(checkbox_todos)
    display(vbox_nomes)
    bairro_selecionado = dropdown_bairro.value
    nomes_selecionados = [cb.description for cb in checkboxes if cb.value]
    df_tabela = atualizar_tabela(bairro_selecionado, nomes_selecionados)
    html_table = df_tabela.to_html(index=False, classes='table table-striped')  # Converte para HTML sem o índice
    display(HTML(html_table))

# Função para marcar ou desmarcar todos os nomes
def selecionar_todos(change):
    for cb in checkboxes:
        cb.value = change['new']

# Criar o widget de dropdown para bairros
bairros_unicos = ['todos'] + df['Bairro'].dropna().unique().tolist()  # Inclui 'todos' e valores únicos de 'Bairro'
dropdown_bairro = widgets.Dropdown(
    options=bairros_unicos,
    value='todos',  # Valor inicial do dropdown
    description='Bairro:',
)

# Criar checkboxes para cada nome
nomes_unicos = df['Nome'].dropna().unique().tolist()  # Valores únicos de 'Nome'
checkboxes = [widgets.Checkbox(value=True, description=nome) for nome in nomes_unicos]

# Checkbox para selecionar/desmarcar todos
checkbox_todos = widgets.Checkbox(
    value=True,
    description='Selecionar todos',
)

# Conectar o checkbox "Selecionar todos" para marcar/desmarcar todas as opções
checkbox_todos.observe(selecionar_todos, names='value')

# Organizar os checkboxes em um VBox
vbox_nomes = widgets.VBox(checkboxes, layout=widgets.Layout(height='200px', overflow_y='scroll'))

# Conectar os dropdowns ao callback
dropdown_bairro.observe(atualizar_tabela_callback, names='value')
for cb in checkboxes:
    cb.observe(atualizar_tabela_callback, names='value')

# Exibir os widgets e a tabela inicial
display(dropdown_bairro)
display(checkbox_todos)
display(vbox_nomes)
atualizar_tabela_callback(None)
```
